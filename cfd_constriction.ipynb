```python
import numpy as np
from tube_constriction import tubeConstriction
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.animation import FuncAnimation
from numba import jit
```

```python
re = 500

numSteps = 250000
numSnap = 100;
```

```python
 
width = int(500)
height = int(50)
```

```python
tube = tubeConstriction(re, width, height, height/9)

tube.initialise() 
tube.geometry (1/10, 40, 20) 

# Numbers
#  500/10 = 50; so constriction is 30 < x < 70
#  constriction for y < 10 and y > 40
time = 0
```

```python
while True:
    speed, density = tube.evolve()
    if time%100 == 0:
        print("Iteration %d, elapsed time %.3f, speed = %.3e" % (time, tube.report_time (), time/tube.report_time()))
    if time%numSnap == 0: 
        fig, (ax1, ax2) = plt.subplots(2, sharex=True, sharey=True,  figsize=(40, 20))
        plt.subplot(121)
        plt.imshow(speed.transpose(), cmap=cm.hot)
        
        plt.subplot(122)
        plt.imshow(density.transpose(), cmap=cm.cool)
        plt.xlabel("x")
        plt.ylabel("y")
        plt.title("Cylindrical with constriction, Re = %.3f, iteration %d" % (tube.reynolds, time))
        plt.show()
        plt.close()
    time += 1
    
    if time>numSteps and time%numSnap == 0:
        break
```
